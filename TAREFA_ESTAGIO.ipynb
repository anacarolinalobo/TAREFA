# PARTE 02: FILTRAGEM DOS DADOS

```python
df = df[["Column8", "Column9", "Column10", "Column14", "Column24", "Column25"]]
df
```

Filtro 03: Município de São Paulo

```python
data = df[df['Column14']== "SÃO PAULO"]
data
```

```python
df2 = pd.unique(pd.Series(data['Column10']))          #Encontrei os valores únicos de local de votação
df2 = sorted(df2)                                     
df2
```

```python
print(len(df2))             #Há 148 locais de votação distintos em SP capital
```

```python
nomes = pd.unique(pd.Series(data['Column24']))          
nomes = sorted(nomes)
nomes
```

```python
for n in nomes:
    data1 = data[data["Column24"] == n]
    d = pd.unique(pd.Series(data1['Column10']))
    if len(d)==148:
        print(n, len(d))
```

```python
data1 = data[data["Column24"] == "TRIPOLI"]                   #Exemplo do candidato TRIPOLI
data1
```

```python
d = pd.unique(pd.Series(data1['Column10']))
d = sorted(d)
d
```

```python
print(len(d))                              #O candidato TRIPOLI obteve votos de 147 locais distintos
```
